```python
with open('/home/haeyoung/korean/phone_recog.nia_snu_hackathon.txt', 'r') as f:
    data1 = f.readlines()

with open('/home/haeyoung/korean/phone_recog.nia_snu_hackathon_set2.txt', 'r') as f:
    data2 = f.readlines()
    
data = data1 + data2
```

```python
info = []
ref = []
ans = []
rec = []    
```

```python
for line in data:
    text = line.split('\n')[0]
    if text.startswith('Aligned transcription') :
        info.append(text)
    elif text.startswith('LAB_WORD') : 
        text = text.split('LAB_WORD: ')[1]
        ref.append(text)
    elif text.startswith('LAB') :
        text = text.split('LAB: ')[1]
        ans.append(text)
    elif text.startswith('REC'):
        text = text.split('REC: ')[1]
        rec.append(text)
```

```python
for sample in info :
    general = sample.split('/')[5]
    nationality.append(sample.split('/')[2])
    speaker.append(sample.split('/')[3])
    sex.append(general.split('-')[1])
    age.append(general.split('-')[2])
    level.append(general.split('-')[4])
    code.append(general.split('-')[5])
    file.append(general.split('-')[6].split('.')[0])
```

```python
result['nationality'] = nationality
result['speaker'] = speaker
result['sex'] = sex
result['age'] = age
result['level'] = level
result['code'] = code
result['file'] = file
result['ref'] = ref
result['ans'] = ans
result['rec'] = rec
```

```python
from jiwer import cer

cer_scores = []
for i in range(len(ans)):
    hypothesis = result['rec'][i]
    ground_truth = result['ans'][i]
    error = cer(ground_truth, hypothesis)
    cer_scores.append(error)
    
print(len(cer_scores))
```

```python
total = 0
for cer in cer_scores:
    total += cer

avg = total / (len(cer_scores))
print('average per score: ', avg)
```

```python
result['per'] = cer_scores
result
```

```python
import re
from collections import Counter
import json

with open('/home/haeyoung/korean/per_utt_results', 'r') as f: 
    lines = f.readlines()

num_utt = int(len(lines)/4)
count = []
C = []
S = []
I = []
D = []
error = []

for num in range(num_utt):
    errors = []
    ref_idx = (num*4)
    hyp_idx = (num*4) + 1
    err_idx = (num*4) + 2
    
    ref = lines[ref_idx]
    hyp = lines[hyp_idx]
    err = lines[err_idx]
    
    ref = re.sub(r'\s+', ' ', ref)
    hyp = re.sub(r'\s+', ' ', hyp)
    err = re.sub(r'\s+', ' ', err)
    #print('\n', ref, '\n', hyp, '\n', err)

    ref = ref.split(' ')[2:]
    hyp = hyp.split(' ')[2:]
    err = err.split(' ')[2:]
    
    count_err = err
    if '' in count_err:
        count_err.remove('')
    else:
        pass
    count_err = Counter(count_err)
    
    cor = count_err['C']
    sub = count_err['S']
    ins = count_err['I']
    dele = count_err['D']
    
    count.append(count_err)
    
    C.append(cor)
    S.append(sub)
    I.append(ins)
    D.append(dele)
    
    
    for j in range(len(ref)):
        if ref[j] == hyp[j]: pass
        else:
            if '*' in hyp[j] :
                res = "deletion of " + "< " + ref[j] + " >"
                errors.append(res)
            elif '*' in ref[j]: 
                res = "insertion of " + "< " + hyp[j] + " >"
                errors.append(res)
            else: 
                res = "substitution of " + "< " + ref[j] + " > to < " + hyp[j] + " >"
                errors.append(res)
                
        final_error = ', '.join(errors)

        
        if len(final_error) == 0:
            final_error = 'none'
            
    error.append(final_error)
        
```

```python
result['count'] = count
result['error'] = error
result['C'] = C
result['S'] = S
result['I'] = I
result['D'] = D
```
