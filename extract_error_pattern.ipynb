```python
import re
from collections import Counter
import json

with open('/home/haeyoung/korean/english/per_utt_results', 'r') as f: 
    lines = f.readlines()

num_utt = int(len(lines)/4)

errors = []
for num in range(num_utt):
    ref_idx = (num*4)
    hyp_idx = (num*4) + 1
    err_idx = (num*4) + 2
    
    ref = lines[ref_idx]
    hyp = lines[hyp_idx]
    err = lines[err_idx]
    
    ref = re.sub(r'\s+', ' ', ref)
    hyp = re.sub(r'\s+', ' ', hyp)
    err = re.sub(r'\s+', ' ', err)
    print('\n', ref, '\n', hyp, '\n', err)

    ref = ref.split(' ')[2:]
    hyp = hyp.split(' ')[2:]
    err = err.split(' ')[2:]
    
    for j in range(len(ref)):
        if ref[j] == hyp[j]: pass
        else:
            if '*' in hyp[j] :
                result = "deletion of " + "< " + ref[j] + " >"
                print(result)
                errors.append(result)
            elif '*' in ref[j]: 
                result = "insertion of " + "< " + hyp[j] + " >"
                print(result)
                errors.append(result)
            else: 
                result = "substitution of " + "< " + ref[j] + " > to < " + hyp[j] + " >"
                print(result)
                errors.append(result)

```

```python

error_dic = Counter(errors)

with open('/home/haeyoung/korean/english/error_patterns_dic', 'w') as f: 
    json.dump(error_dic, f, indent=4)
    print(error_dic)
```

```python
sorted_tuple = sorted(error_dic.items(), key = lambda item: item[1], reverse = True)
print(sorted_tuple[:50])
with open('/home/haeyoung/korean/english/error_patterns_dic_50', 'w') as f: 
    f.write('\n'.join(f'{tup[0]} {tup[1]}' for tup in sorted_tuple))
```
